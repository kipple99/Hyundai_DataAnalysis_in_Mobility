```python
iris_df = sns.load_dataset('iris')
```

```python
plt.figure()

sns.pairplot(iris_df, hue='species')

plt.show()
```

```python
# KNN(K Nearest Neighbors)
# neighbors => 거리
# 새로운 데이터 / 기존 데이터와 거리를 계산
# 거리가 짧은거 n개를 살펴봄
# 이 데이터들의 target을 살펴봐서 다수결
```

```python
# 사아킷런 사용법
# 1. 모델을 생성
# 2. 모델을 훈련
# 3. 예측, 검정
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# 1. 모델을 생성

knn = KNeighborsClassifier()
```

```python
# 2. 모델을 훈련
# feature와 target을 분리
# target은 무조건 1차원
# feature는 무조건 2차원

y = iris_df['species']
X = iris_df.iloc[ : , : -1 ]

knn.fit(X, y)
```

```python
# 3. 예측, 검정

knn.predict([[5.0, 3.3, 1.45, 0.15]])
```

```python
# 훈련용 데이터와 검증용 데이터로 나누기
# 8대 2로 나누기
# 섞어서 나눠야 함.(그럴줄알고 사이킷 런은 다 준비해놨어)

from sklearn.model_selection import train_test_split
```

```python
train_x, valid_x, train_y, valid_y = train_test_split(X, y, 
                 test_size=0.2, 
                 random_state=77, 
                 stratify=y) # random_state : 재현성(맞춰주기)
```

```python
train_x
```

```python
# 1. 모델을 생성
knn = KNeighborsClassifier()

# 2. 모델을 훈련(fit)
knn.fit(train_x, train_y)

# 3. 예측, 검증(socre)
knn.predict(valid_x) == valid_y # 1
knn.score(valid_x, valid_y) # 2
```

```python
# 1. 하이퍼 파라미터 세팅
# 2. 모델을 생성
# 3. 모델을 훈련
# 4. 예측, 검증
```

```python
# 교차 검증 => 적은 데이터를 n등분을 해서 그 중 에서 훈련 검증을 교차로 하는 것
# 마지막으로 평균내고 합친 것 모든데이터 훈련 / 모든 데이터 검증
```

```python
from sklearn.model_selection import cross_val_score
```

```python
knn_1 = KNeighborsClassifier(n_neighbors=1)

cross_val_score(knn_1, train_x, train_y, cv=4).mean()
```

```python
knn_5 = KNeighborsClassifier(n_neighbors=5)

cross_val_score(knn_5, train_x, train_y, cv=4).mean()
```

```python
# 요령이 있음. 어렴풋이 골라야함. 정확히 딱 정해지지가 않음
# k가 1일때, 주변에 1개 데이터만 보는것 => 새로운 에측값이 위치가 살짝만 바뀌어도 불안정함(오버피팅)
# 반대로 k를 무한대로 설정한다면? => 그냥 제일 많은 빈도의 데이터를 예측값으로 줌(=평균값으로 보여주는 것 뿐임)(언더피팅)
# 하이퍼 파라미터는 오버와 언더 피팅 그 사이에 적절한 선을 찾는 것

# 어떻게 찾을 것이냐?
# 변곡점 찾기, 이값을 정확히 찾는 것은 불가능 함. 여러가지 데이터를 넣어서 올라가다 떨어지는 부분을 찾는 것임. 교차검증을 하면서 => 엘보우 메소드(사람 팔꿈치모양 그래프)
```

```python
# 1~30까지 바꿔가며 교차검증을 해보자

score_list = []

for k in range(1, 31):
    knn = KNeighborsClassifier(n_neighbors = k)
    score = cross_val_score(knn, train_x, train_y, cv=4).mean()
    score_list.append(score)
```

```python
plt.figure()

plt.plot(range(1, 31), score_list)

plt.show()
```
